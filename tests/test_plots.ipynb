```python
def plot_cos(phi):
    x = np.linspace(0, 1.0, 501)
    y = np.cos(2 * np.pi * x + phi)
    plt.plot(x, y)

phis = np.linspace(0, 2 * np.pi, 101)
plot_cos_gif = gif_it(plot_cos, fps=25, dpi=100)

plot_cos_gif(phis)
```
